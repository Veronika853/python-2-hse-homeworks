```python
# импорт библиотек
from bs4 import BeautifulSoup
import requests
```

```python
# url адрес страницы
url = "https://adme.media/svoboda-psihologiya/100-voprosov-kotorye-nuzhno-obsudit-s-rebenkom-chtoby-ukrepit-s-nim-svyaz-2043665/"
```

```python
# выполняем GET запрос
response = requests.get(url)
```

```python
# создаем объект BeautifulSoup с текстом страницы, парсер html.parser
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# извлекаем все содержимое тегов div с классом e599d42b92530cddc130 (блоки с вопросами)
asks = soup.find_all('div', class_="e599d42b92530cddc130")
```

```python
# извлекаем названия блоков с помощью фильтрации (в начале цифра)
title_tags = list(filter(lambda x: x[1].text[0].isdigit(), enumerate(asks)))
```

```python
# из части с названиями блоков забираем только текст названий
titles = [tag.text[tag.text.find('.') + 2:].replace('\xa0', ' ') for _, tag in title_tags]
```

```python
# извлекаем текст страницы с вопросами для каждого блока
question_tags  = [asks[i + 1] for i, _ in title_tags]
```

```python
# извлекаем только текст вопросов
questions = [[p_tag.text.replace('\xa0', ' ') for p_tag in tag.find_all('p')] for tag in question_tags]
```

```python
# запаковываем названия блоков вместе с вопросами из них
result = dict(zip(titles, questions))
```
